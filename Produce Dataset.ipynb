```python
## there are 118 different sub-categories in this dataset, which will be too difficult to analyze
## grouping them into more general categories would be suggested for further analysis

df['Category'].value_counts()
```

```python
## displaying an array of the Category column

df['Category'].unique()
```

```python
## instantiate an empty list to store each sub-category into 'general' category, copied and revised code

general_category_list = []

## converting sub-categories into a list

sub_category = df['Category'].drop_duplicates(inplace=False).tolist()

## need a for-loop to loop through each sub-category and placing them into each specifc general category based on 
## the positioning of the category in the above array

for x in df['Category'].tolist():
    if x in sub_category[:7]:
        general_category_list.append("Fruits & Vegetables")
    elif x in sub_category[7:16]:
        general_category_list.append("Dairy & Eggs")
    elif x in sub_category[16:29]:
        general_category_list.append("Pantry")
    elif x in sub_category[29:50]:
        general_category_list.append("International Foods")
    elif x in sub_category[50:57]:
        general_category_list.append("Meat")
    elif x in sub_category[57:64]:
        general_category_list.append("Fish & Seafood")
    elif x in sub_category[64:68]:
        general_category_list.append("Snacks, Chips & Candy")
    elif x in sub_category[68:77]:
        general_category_list.append("Beverages")
    elif x in sub_category[77:86]:
        general_category_list.append("Frozen Foods")
    elif x in sub_category[86:95]:
        general_category_list.append('Natural & Organic Foods')
    elif x in sub_category[95:102]:
        general_category_list.append('Deli')
    elif x in sub_category[102:108]:
        general_category_list.append('Bakery')
    elif x in sub_category[108:]:
        general_category_list.append('Prepared Meals')   
```

```python
df['General Category'] = general_category_list
```

```python
df.head()
```

```python
## checking total categories from General Category column

df['General Category'].value_counts()
```

```python
## calculating price difference between each product

## cannot subtract 'Previous Price' from 'Current Price' because it is a string, need to convert to float

df['Current Price'] - df['Previous Price']
```

```python
## could not parse string into float without removing the ',' from string
## removed the ',' and now converted string dtype to numeric

df['Current Price'] = df['Current Price'].str.replace(",", "")

df['Current Price'] = df['Current Price'].apply(pd.to_numeric)
```

```python
df['Previous Price'] = df['Previous Price'].str.replace(",", "")

df['Previous Price'] = df['Previous Price'].apply(pd.to_numeric)
```

```python
df.dtypes
```

```python
df['Price Difference'] = df['Current Price'] - df['Previous Price']

df.head()
```

```python
## removing $$ signs 

df['Price Per Each'].str.replace('$', '')
```

```python
## removing the units of measure string behind the price in column 'Price Per Each'

df['Price Per Each'] = df['Price Per Each'].str.extract('(\d+\.\d+)').astype(float)
```

```python
## converting 'Amount' to int datatype

df['Amount'] = df['Amount'].str.extract('(\d+)').astype(int)
```

```python
## calculating 'total sales' by multiplying 'Price per Each' by 'Amount'

df['Total Sales'] = df.groupby('Name')['Price Per Each'].transform('first') * df['Amount']
```

```python
df.head()
```

```python
df.dtypes
```

```python
## calculating 'Current Price' Percent Increase/Decrease comparing to 'Previous Price' 
## by taking the 'Price Difference' column created from above and divide by the 'Previous Price', multiplied by 100

df['Percent Increase/Decrease'] = ((df['Price Difference'] / df['Previous Price']) * 100).round(2)
```

```python
df.head()
```

```python
## bar graph to show grouped 'Total Sales' per each 'General Category' categories

grouped_data = df.groupby('General Category')['Total Sales'].sum()

plt.figure(figsize=(12, 6))
grouped_data.plot(kind='bar', color='purple')
plt.title('Total Sales by Category')
plt.xlabel('Category Group')
plt.ylabel('Total Sales(Million)')
plt.show()
```

```python
## As the graph shows, 'Beverages' , 'Deli', and 'Pantry' appears to have the most total sales amongst the general categories 
## with each surpassing $1.6 million in total sales. The 'Fish & Seafood' and 'Fruits & Vegetables' categories appear to have 
## the least amount of total sales. This could mean that there were more products/quantity of products or higher prices per quanity 
## in the 'Beverages', 'Deli', and 'Pantry' categories thus resulting in more $ sales, whereas less products/quantity of products or
## lower prices in 'Fish & Seafood' and 'Fruits & Vegetables' categories. 
## A deeper analysis may provide a more useful insight on this dataset.
```

```python
## bar graph to show total price differences in each category

grouped_data2 = df.groupby('General Category')['Price Difference'].sum()


plt.figure(figsize=(12,6))
grouped_data2.plot(kind='bar', color='violet')
plt.title('Total Price Difference by Category')
plt.xlabel('Category Group')
plt.ylabel('Total Price Difference')
plt.show()
```

```python
## This graph shows there is a big price difference amongst 'Pantry' items with a total cumalative price difference of over 250$.
## 'Bakery', 'Dairy & Eggs', 'Natural & Organic Foods', and 'Prepared Meals' appears to show the lease amount of price differences
## between current and old prices. Another analysis may be needed to see if price difference in items of each respective
## category influcences total categorical sales. 
```

```python
## bar graph to show average price differences in each category

grouped_data1 = df.groupby('General Category')['Percent Increase/Decrease'].mean()


plt.figure(figsize=(12,6))
grouped_data1.plot(kind='bar', color='red')
plt.title('Average Percent Increase/Decrease by Category')
plt.xlabel('Category Group')
plt.ylabel('Average Percent Increase/Decrease')
plt.show()
```

```python
## 'International Foods' appear to have the largest percent decrease in comparison to the other categories with nearly 3.3% 
## decrease. 'Dairy & Eggs' , 'Meat', and 'Prepared Meals' all appear to have less than .5%, so it would mean that prices
## amongst these categories did not change much. 
```
